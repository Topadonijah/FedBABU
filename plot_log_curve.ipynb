```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Unbalanced IID

```python
dataset = 'cifar100'
model = 'mobile'
user_num = 100
frac = 1.0
local_ep = 4

shard_per_user = 2
bu = 5
md = 50

body_body = './save/{}/{}_iidTrue_num{}_C{}_le{}/shard{}_unbalanced_bu{}_md{}/run1/local_upt_body_aggr_body/results.csv'.format(dataset, model, user_num, frac, local_ep, shard_per_user, bu, md)
full_full = './save/{}/{}_iidTrue_num{}_C{}_le{}/shard{}_unbalanced_bu{}_md{}/run1/local_upt_full_aggr_full/results.csv'.format(dataset, model, user_num, frac, local_ep, shard_per_user, bu, md)

body_body = pd.read_csv(body_body)['acc_test']
full_full = pd.read_csv(full_full)['acc_test']

x = range(local_ep, 321, local_ep)
plt.plot(x, body_body, marker='o', markersize=3, label='body_body')
plt.plot(x, full_full, marker='o', markersize=3, label='full_full (FedAvg)')

plt.title('users #: {}, user fraction: {},\nlocal epoch(s): {}, classes # per user: {}'.format(user_num, frac, local_ep, shard_per_user))
plt.legend()
plt.show()
plt.close()
```

# Balanced nonIID~IID

```python
dataset = 'cifar100'
model = 'mobile'
user_num = 100
frac = 1.0
local_ep = 4

shard_per_user = 2

body_body = './save/{}/{}_iidFalse_num{}_C{}_le{}/shard{}/run1/local_upt_body_aggr_body/results.csv'.format(dataset, model, user_num, frac, local_ep, shard_per_user)
head_head = './save/{}/{}_iidFalse_num{}_C{}_le{}/shard{}/run1/local_upt_head_aggr_head/results.csv'.format(dataset, model, user_num, frac, local_ep, shard_per_user)
full_body = './save/{}/{}_iidFalse_num{}_C{}_le{}/shard{}/run1/local_upt_full_aggr_body/results.csv'.format(dataset, model, user_num, frac, local_ep, shard_per_user)
full_head = './save/{}/{}_iidFalse_num{}_C{}_le{}/shard{}/run1/local_upt_full_aggr_head/results.csv'.format(dataset, model, user_num, frac, local_ep, shard_per_user)
full_full = './save/{}/{}_iidFalse_num{}_C{}_le{}/shard{}/run1/local_upt_full_aggr_full/results.csv'.format(dataset, model, user_num, frac, local_ep, shard_per_user)

body_body = pd.read_csv(body_body)['acc_test']
head_head = pd.read_csv(head_head)['acc_test']
full_body = pd.read_csv(full_body)['acc_test']
full_head = pd.read_csv(full_head)['acc_test']
full_full = pd.read_csv(full_full)['acc_test']

x = range(local_ep, 321, local_ep)
plt.plot(x, body_body, marker='o', markersize=3, label='body_body')
plt.plot(x, head_head, marker='o', markersize=3, label='head_head')
plt.plot(x, full_body, marker='o', markersize=3, label='full_body')
plt.plot(x, full_head, marker='o', markersize=3, label='full_head')
plt.plot(x, full_full, marker='o', markersize=3, label='full_full (FedAvg)')

plt.title('users #: {}, user fraction: {},\nlocal epoch(s): {}, classes # per user: {}'.format(user_num, frac, local_ep, shard_per_user))
plt.legend()
plt.show()
plt.close()
```
